```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
sheets = pd.read_excel('I Know First Company Names.xlsx')
```

```python
name = str(input("Enter Company Ticker Name: " + "\n"))
```

```python
def desc (name):
    url = 'https://www.reuters.com/finance/stocks/company-profile/' + name
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html)
    soup1 = soup.findAll('p')[0].text
    soup4 = soup.findAll('h1')
    name = soup4[0].text
    name1 = name.partition(' ')[2]
    full_name = name1.rsplit(' ', 1)[0]
    first, *middle, last = name1.split()
    ticker = last[1:-1]
    if len(soup1) < 125:
        soup1 = soup.findAll('div')
        soup2= soup1[101].text[31:]
        sep ='Â»'
        head = soup2.split(sep, 1)[0]
        return [full_name, ticker, head]
    else:
        return[full_name, ticker, soup1]
w = pd.DataFrame(np.array(desc(name)).reshape(-1,len(desc(name))))
w.columns = ['Name', 'Ticker Symbol', 'Explanation']
w
```

```python
df = pd.concat([sheets, w])
```
